_Centrality_ is a term to describe importance of individual nodes in a graph. \[a\]
There has been a lot of research carried out in this topic for network analysis to answer the question,
"Which are the most important nodes (vertices) in a graph?"
Here is the list of different metrics to determine that, which I would like to discuss:

```python
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def draw(G, pos, measures, measure_name):

    nodes = nx.draw_networkx_nodes(G,pos, node_color=measures.values(), nodelist=measures.keys(), cmap='plasma', with_labels=False)
    edges = nx.draw_networkx_edges(G,pos)
    plt.title(measure_name)
    plt.colorbar(nodes)
    plt.axis('off')
    plt.show()
```
